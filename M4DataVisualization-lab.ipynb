## Visualizing distribution of data


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Create a cursor object
cursor = conn.cursor()

# Query to list all tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")

# Fetch all results
tables = cursor.fetchall()

# Print the list of tables
print("Tables in the database:")
for table in tables:
    print(table[0])

# Close the connection
conn.close()
```

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Create a cursor object
cursor = conn.cursor()

# Query to list all columns in the master table
cursor.execute("PRAGMA table_info(master);")

# Fetch all results
columns = cursor.fetchall()

# Print the list of columns
print("Columns in the 'master' table:")
for column in columns:
    print(column[1])

# Close the connection
conn.close()
```

### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to read data from the master table
query = "SELECT ConvertedComp FROM master"

# Read the data into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Plot the histogram for the column ConvertedComp
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, edgecolor='k', alpha=0.7)
plt.title('Distribution of Converted Compensation')
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to read Age and WorkWeekHrs columns from the master table
query = "SELECT Age, WorkWeekHrs FROM master"

# Read the data into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Drop rows with missing values in either Age or WorkWeekHrs
df = df.dropna(subset=['Age', 'WorkWeekHrs'])

# Plot the scatter plot for Age and WorkWeekHrs
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to read WorkWeekHrs, CodeRevHrs, and Age columns from the master table
query = "SELECT WorkWeekHrs, CodeRevHrs, Age FROM master"

# Read the data into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Drop rows with missing values in WorkWeekHrs, CodeRevHrs, or Age
df = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

# Plot the bubble plot
plt.figure(figsize=(10, 6))
bubble_size = df['Age'] * 10  # Scale bubble size appropriately
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_size, alpha=0.5, c=df['Age'], cmap='viridis')
plt.colorbar(label='Age')
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs (Bubble Size by Age)')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the counts of each database that respondents wish to learn next year
query = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

# Read the data into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Plot the pie chart for the top 5 databases
plt.figure(figsize=(10, 6))
plt.pie(df['count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to read WorkWeekHrs, CodeRevHrs, and Age columns from the master table for the age group 30 to 35
query = """
SELECT WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""

# Read the data into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Calculate the median WorkWeekHrs and CodeRevHrs for the specified age group
median_work_week_hrs = df['WorkWeekHrs'].median()
median_code_rev_hrs = df['CodeRevHrs'].median()

# Create a DataFrame for plotting
data = pd.DataFrame({
    'Hours': [median_work_week_hrs, median_code_rev_hrs]
}, index=['WorkWeekHrs', 'CodeRevHrs'])

# Plot the stacked bar chart
data.T.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Median Hours')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to read Age and ConvertedComp columns from the master table for ages 45 to 60
query = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""

# Read the data into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Calculate the median ConvertedComp for each age within the range
median_converted_comp = df.groupby('Age')['ConvertedComp'].median()

# Convert the index to integers
median_converted_comp.index = median_converted_comp.index.astype(int)

# Plot the median ConvertedComp for each age as a line chart
plt.figure(figsize=(10, 6))
median_converted_comp.plot(kind='line', marker='o')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp (USD)')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the counts of each MainBranch category
query = """
SELECT MainBranch, COUNT(*) as count
FROM master
GROUP BY MainBranch
ORDER BY count DESC
"""

# Read the data into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Increase figure size for better readability
plt.figure(figsize=(12, 8))

# Plot the horizontal bar chart for MainBranch
plt.barh(df['MainBranch'], df['count'], color='skyblue')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Distribution of MainBranch')
plt.grid(True)

# Ensure y-axis labels are fully visible
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the counts of each language respondents wish to learn next year
query = """
SELECT LanguageDesireNextYear, COUNT(*) as count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY count DESC
"""

# Read the data into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Display the full list of languages and their counts
print(df)

# Check if Python is in the DataFrame
if 'Python' in df['LanguageDesireNextYear'].values:
    # Find the rank of Python
    df['Rank'] = df['count'].rank(ascending=False, method='min')
    python_rank = df[df['LanguageDesireNextYear'] == 'Python']['Rank'].values[0]
    python_rank = int(python_rank)  # Convert rank to integer
    print(f"The rank of Python is: {python_rank}")
else:
    print("Python is not in the list of languages respondents wish to learn next year.")
```

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the count of respondents who currently work with SQL
query = """
SELECT COUNT(*) as count
FROM LanguageWorkedWith
WHERE LanguageWorkedWith LIKE '%SQL%'
"""

# Read the result into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Get the count of respondents who work with SQL
sql_count = df['count'].values[0]

# Print the result
print(f"The number of respondents who currently work with SQL is: {sql_count}")
```

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get all entries in DatabaseWorkedWith
query = "SELECT Respondent, DatabaseWorkedWith FROM DatabaseWorkedWith"

# Read the data into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Group by Respondent and aggregate the databases they work with
grouped = df.groupby('Respondent')['DatabaseWorkedWith'].apply(lambda x: ','.join(x)).reset_index()

# Filter out respondents who work only with MySQL
mysql_only_count = grouped[grouped['DatabaseWorkedWith'] == 'MySQL'].shape[0]

# Print the result
print(f"The number of respondents who currently work only with MySQL is: {mysql_only_count}")
```
